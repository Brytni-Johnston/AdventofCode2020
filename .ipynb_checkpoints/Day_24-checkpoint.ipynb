```python
import pandas as pd
import numpy as np
import networkx as nx

moves = pd.read_csv('day24input.txt',header=None)
```

```python
#Part 1: How many tiles are black after all tiles have been flipped?

Dir = ['e','w','ne','nw','se','sw']
x = {'e':1,'w':-1,'ne':1/2,'nw':-1/2,'se':1/2,'sw':-1/2}
y = {'e':0,'w':0,'ne':np.sqrt(3)/2,'nw':np.sqrt(3)/2,'se':-1*np.sqrt(3)/2,'sw':-1*np.sqrt(3)/2}
Tiles = {}

for i in moves.index:
    loc_x = 0
    loc_y = 0
    while len(moves.loc[i,0]) > 0:
        for j in Dir:
            if moves.loc[i,0].startswith(j):
                loc_x += x[j]
                loc_y += y[j]
                moves.loc[i,0] = moves.loc[i,0].replace(j,'',1)
    loc_y = np.round(loc_y,5)
    if (loc_x,loc_y) in Tiles.keys():
        Tiles[(loc_x,loc_y)] = not Tiles[(loc_x,loc_y)]
    else:
        Tiles[(loc_x,loc_y)] = True

Total = 0
for i in Tiles.keys():
    if Tiles[i]:
        Total += 1
        
print(Total)
```
